```python
#Since there is 4 coloumns having Missing values so , Lets impute it

from sklearn.impute import SimpleImputer

imputer1= SimpleImputer(strategy='most_frequent')
df['name']=imputer1.fit_transform(df[['name']])
```

```python
df.isnull().sum()
```

```python
imputer2= SimpleImputer(strategy='most_frequent')
df['host_name']=imputer2.fit_transform(df[['host_name']])

imputer3= SimpleImputer(strategy='most_frequent')
df['reviews_per_month']=imputer3.fit_transform(df[['reviews_per_month']])
```

```python
df['last_review'] = df['last_review'].fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
#Check Duplicates

duplicate_rows = df.duplicated()
print(f"Number of duplicate rows: {duplicate_rows.sum()}")
```

```python
#Since there is no duplicates, so we will move towards Standardization

from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the numeric columns
numeric_cols = ['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Standardize the date column (last_review)
df['last_review'] = pd.to_datetime(df['last_review'])

# Standardize the categorical columns (if needed)
categorical_cols = ['name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type']
for col in categorical_cols:
    df[col] = df[col].astype('category')
    
```

```python
def handle_outliers_zscore(df, threshold=3):
    # Calculate the mean and standard deviation
    mean = np.mean(df)
    std = np.std(df)
    
    # Calculate Z-scores for each data point
    z_scores = np.abs((df - mean) / std)
    
    # Identify outliers based on the threshold
    outliers = df[z_scores > threshold]
    
    # Replace outliers with NaN values
    df[z_scores > threshold] = np.nan
    
    return df
```

```python
# Handle outliers using Z-score method
data_handled = handle_outliers_zscore(df)
```
