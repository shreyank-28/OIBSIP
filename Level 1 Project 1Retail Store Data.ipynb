```python
from statsmodels.tsa.seasonal import seasonal_decompose
# Group the data by date and calculate the total sales amount
sales_by_date = df.groupby('Date')['Sales_Amount'].sum().reset_index()

# Create a time series object
sales_ts = pd.Series(sales_by_date['Sales_Amount'].values, index=sales_by_date['Date'])

# Perform seasonal decomposition
result = seasonal_decompose(sales_ts, model='additive', period=30)

# Plot the decomposition
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(sales_ts)
plt.title('Original Series')
plt.subplot(412)
plt.plot(result.trend)
plt.title('Trend')
plt.subplot(413)
plt.plot(result.seasonal)
plt.title('Seasonality')
plt.subplot(414)
plt.plot(result.resid)
plt.title('Residuals')
plt.tight_layout()
plt.show()
```

```python
# Calculate the year-over-year growth rate
sales_by_date['YoY_Growth'] = sales_by_date['Sales_Amount'].pct_change(periods=365)

# Print the results
print(sales_by_date)
```

```python
top_SKUS=df.groupby('SKU')['Sales_Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
top_SKUS.plot(kind='bar')
plt.xlabel=('SKU')
plt.ylabel=('Total Sales Amount')
plt.show()
```
